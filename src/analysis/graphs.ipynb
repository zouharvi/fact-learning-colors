```python
import matplotlib.pyplot as plt
import numpy as np
fig = plt.figure(figsize=(5,3))
ax = fig.add_axes([0,0,1,1])
x = np.arange(2)
group = ['Control','Random','Difficulty']
y=[89.47,77.55,90.0]
x_pos=[0,0.5,1]
ax.bar(x_pos[0],y[0],color='#FFA500', width = 0.25, align="edge")
ax.bar(x_pos[1],y[1],color='#7BC8F6', width = 0.25, align="edge")
ax.bar(x_pos[2],y[2],color='#FF6347', width = 0.25, align="edge")
plt.xticks([x+0.13 for x in x_pos], group)
plt.title("PT0")
plt.text(0.05, y[0], str(y[0]-5)+"%")
plt.text(0.56, y[1], str(y[1]-5)+"%")
plt.text(1.05, y[2], str(y[2]-5)+"%")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
fig = plt.figure(figsize=(5,3))
ax = fig.add_axes([0,0,1,1])
x = np.arange(2)
group = ['Control','Random','Difficulty']
y=[88.72,79.59,84.12]
ax.bar(group[0],y[0],color='yellow', width = 0.25)
ax.bar(group[1],y[1],color='g', width = 0.25)
ax.bar(group[2],y[2],color='r', width = 0.25)
plt.title("PT1")
for index, value in enumerate(y):
    plt.text(index, value, str(value)+"%")
plt.tight_layout()
plt.show()
```
